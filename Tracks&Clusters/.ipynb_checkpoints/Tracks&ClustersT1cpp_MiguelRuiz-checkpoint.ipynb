```c++
#include <iostream>
#include <string>
#include <stdio.h>
```

```c++
//TFile *file = TFile::Open("Tracks_Clusters.root");
TFile *file = TFile::Open("https://csandova.web.cern.ch/csandova/HEP-Ex-Course/Code/TracksClusters/Tracks_Clusters.root");
```

```c++
TTree *tree = (TTree*) file->Get("JetRecoTree");
nentries = (Int_t)tree->GetEntries();
int nbytes = 0;
tree->Print()
```

```c++
UInt_t npv = -1;
float mu_average = -1;
vector<float> *tracks_pt;
vector<float> *tracks_eta;
vector<float> *tracks_phi;
vector<float> *tracks_m;
vector<float> *tracks_vtx;
vector<float> *clusters_pt;
vector<float> *clusters_eta;
vector<float> *clusters_phi;
vector<float> *clusters_m;

tree->SetBranchAddress("NPV", &npv);
tree->SetBranchAddress("mu_average", &mu_average);
tree->SetBranchAddress("Tracks_pt", &tracks_pt);
tree->SetBranchAddress("Tracks_eta", &tracks_eta);
tree->SetBranchAddress("Tracks_phi", &tracks_phi);
tree->SetBranchAddress("Tracks_m", &tracks_m);
tree->SetBranchAddress("Tracks_vtx", &tracks_vtx);
tree->SetBranchAddress("Clusters_pt", &clusters_pt);
tree->SetBranchAddress("Clusters_eta", &clusters_eta);
tree->SetBranchAddress("Clusters_phi", &clusters_phi);
tree->SetBranchAddress("Clusters_m", &clusters_m);
```

```c++
TCanvas *canvas = new TCanvas("Canvas","a first way to plot a variable",800,600);
```

```c++
TH1F *hist_npv = new TH1F("NPV","Example plot: Number of primary vertices; NPV ; Events ",50,1,50);
TH1F *hist_mu_average = new TH1F("mu_average","Example plot: Number of average interactions per bunch-crossing; mu average ; Events ",100,1,100);
TH2F *hist2d_npv_mu = new TH2F("2DHist_npv_mu", "NPV vs mu 2d Histogram; mu_average; NPV", 100, 1, 100, 50, 1, 50);
TH1F *hist_nTrks = new TH1F("NTracks","Example plot: Number of tracks; NTracks ; Events ",500,0,2000);
TH1F *hist_nCltrs = new TH1F("NClusters","Example plot: Number of clusters; NClusters ; Events ",500,0,2000);
TH2F *hist2d_npv_mu_ntr_ncl = new TH2F("2DHist_npv_mu_ntr_ncl", "NPV and mu vs nTracks and nClusters 2d Histogram; nTracks and nClusters; NPV and mu_average", 500, 0, 2000, 100, 1, 100);
TH1F *hist_lead_track_pT = new TH1F("Lead_Track_pT","Example plot: Track pT; tracks_pT (MeV) ; Events ",50,100,4000);
TH1F *hist_track_eta = new TH1F("Track_eta","Example plot: Track eta; tracks_eta ; Events ",100,-3,3);
TH1F *hist_track_phi = new TH1F("Track_phi","Example plot: Track phi; tracks_phi ; Events ",100,-4, 4);
TH1F *hist_track_m = new TH1F("Track_m","Example plot: Track m; tracks_m ; Events ",100,-10000,10000);
TH1F *hist_track_vtx = new TH1F("Track_vtx","Tracks_vtx; tracks_vtx ; Events ",120,-20,100);
TH1F *hist_clusters_pt = new TH1F("Clusters_pt","Clusters_pt; clusters_pT (MeV); Events ",1000,0,50000);
TH1F *hist_clusters_eta = new TH1F("Clusters_eta","Clusters_eta; clusters_eta ; Events ",100,-5, 5);
TH1F *hist_clusters_phi = new TH1F("Clusters_phi","Clusters_phi; clusters_phi ; Events ",200,-4, 4);
TH1F *hist_clusters_m = new TH1F("Clusters_m","Clusters_m; clusters_m ; Events ",100,-1.5, 1.5);
```

# Excercise: 1. Plot the number of average interactions per bunch-crossing (mu_average)

```c++
nbytes = 0;

for (int ii= 0; ii< nentries; ii++)
{
    nbytes = tree->GetEntry(ii);
    hist_mu_average->Fill(mu_average);
}

std::cout << "Done!" << std::endl;
```

```c++
hist_mu_average->SetFillColor(kBlue);
hist_mu_average->Draw();
canvas->Draw();
```

```c++
Float_t scale = hist_mu_average->Integral();
hist_mu_average->Scale(1/scale);
hist_mu_average->SetFillColor(2);
hist_mu_average->Draw();
canvas->Draw();
```

# Excersice 2: Make a 2D histogram of NPV vs. mu 

```c++
nbytes = 0;

for (int ii=0; ii<nentries; ii++)
{
    nbytes = tree->GetEntry(ii);
    hist2d_npv_mu->Fill(mu_average, npv);
}

std::cout << "Done!" << std::endl;
```

```c++
hist2d_npv_mu->Draw();
canvas->Draw();
```

# Excercise: 3. Make a 2D histogram of NPV and mu vs. nTracks and nClusters 

```c++
nbytes = 0;

for (int ii = 0; ii < nentries; ii++)
{
    nbytes = tree->GetEntry(ii);
    hist_nTrks->Fill(tracks_pt->size());
    hist_nCltrs->Fill(clusters_pt->size());
}

std::cout << "Done!" << std::endl;
```

```c++
hist_nTrks->SetFillColor(kBlue);
hist_nTrks->Draw();
canvas->Draw();
```

```c++
hist_nCltrs->SetFillColor(kYellow);
hist_nCltrs->Draw();
canvas->Draw();
```

```c++
nbytes = 0;

for (int ii=0; ii<nentries; ii++)
{
    nbytes = tree->GetEntry(ii);
    hist2d_npv_mu_ntr_ncl->Fill(tracks_pt->size(), npv);
    hist2d_npv_mu_ntr_ncl->Fill(clusters_pt->size(), mu_average);    
}

std::cout << "Done!" << std::endl;
```

## Excercise 5: Interpret the meaning of each variable

En cuanto a las variables de los tracks, como por ejemplo el pT, eta, phi, m y vtx hace referencia al momento transverso, la pseudorapidez, el ángulo phi que hace la proyección del track en el plano transverso, m (no estoy seguro, ya que al imprimir los valores de track_m obtengo un valor muy cercano a 139MeV, así que supongo es la masa de la partícula cargada (tal vez un pion) la cual deja los tracks hallados) sería la masa de la partícula cargada que produce el track, y el vtx nos asociaría los tracks que vienen de un mismo vertice.

Para el caso de las variables de los clusters sucedería algo similar, ya que el pT sería el momento transverso asociado a un determinado cluster, eta y phi nos daría la posición en el detector del cluster.

## Excersice 6: What is "Particle PDG ID"? 

El PDG ID para una partícula es el número asociado a ésta partícula de acuerdo con el esquema de numeración establecido por el Particle Data Group (PDG) y permite identificar a ésta partícula de las otras, ya que su asignación es única. Así, para el electrón se le asigna por ejemplo el número 11 mientras que a los muones se les asigna el 13. También se les asigna un número para los quarks, los leptones, bosones, etc y ésta información es útil si queremos determinar el tipo de partículas que son producidas en cada evento.
